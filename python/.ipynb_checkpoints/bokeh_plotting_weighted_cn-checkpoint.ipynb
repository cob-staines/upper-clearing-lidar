```python
import numpy as np
import pandas as pd

import holoviews as hv
from holoviews import dim, opts
hv.extension('bokeh')

import datashader as ds
from datashader import transfer_functions as tf

# load data
data_in ='C:\\Users\\Cob\\index\\educational\\usask\\research\\masters\\data\\lidar\\ray_sampling\\batches\\lrs_hemi_uf_.25m_180px\\outputs\\rshmetalog_weighted_cn.csv'
data = pd.read_csv(data_in)

# filter to test set
data = data.loc[data.training_set.values, :]

plot_width = int(150)
plot_height = int(plot_width//1.2)
zoom_multiplier = 3

def covar_plot(x_var, y_var, x_range, y_range):
    cvs = ds.Canvas(plot_width=plot_width, plot_height=plot_height, x_range=x_range, y_range=y_range)
    agg = cvs.points(data, x_var, y_var, ds.count(y_var))
    return hv.Image(agg).opts(colorbar=True, logz=False, width=int(plot_width*zoom_multiplier), height=int(plot_height*zoom_multiplier))

x_var = ['covariant']
y_var = ['log_cn_abs_weighted', 'log_cn_sqr_weighted']

plot_dict={}
for yy in y_var:
    # y_range = (np.nanmin(data.loc[:, yy]),np.nanmax(data.loc[:, yy]))
    y_range = (np.nanquantile(data.loc[:, yy], .001),np.nanquantile(data.loc[:, yy], .999))
    # y_range = (0, 2)
    for xx in x_var:
        # x_range = (np.nanmin(data.loc[:, x_var]),np.nanmax(data.loc[:, x_var]))
        x_range = (np.nanquantile(data.loc[:, xx], .001),np.nanquantile(data.loc[:, xx], .999))
        # x_range = (0, 200)
        plot_dict[(yy, xx)] = covar_plot(xx, yy, x_range, y_range)


kdims = [hv.Dimension(('y_var', 'yy')),
         hv.Dimension(('x_var', 'xx'))]
holomap = hv.HoloMap(plot_dict, kdims=kdims)
# holomap.opts(opts.Curve(width=600))

grid = hv.GridSpace(holomap)
# grid.opts(
#     opts.GridSpace(plot_size=75),
#     opts.Curve(width=100))

ndlayout = hv.NdLayout(grid)
ndlayout.opts(opts.Image(width=plot_width*zoom_multiplier, height=plot_height*zoom_multiplier)).cols(len(x_var))
ndlayout
# hv.save(ndlayout, 'C:\\Users\\Cob\\index\\educational\\usask\\research\\masters\\graphics\\AGU_presentation\\rs_lai_TRUE_ln_dswe_19_045-19_052.png', fmt='png')
```
